```python
#primera integral
import numpy as np
from scipy.integrate import quad
from scipy.special import roots_legendre

# Definir la función que se va a integrar
def f(x):
    return (1 / np.sqrt(2 * np.pi)) * np.exp(-x**2 / 2)

# Método de la regla del trapecio
def trapecio(f, a, b, n):
    x = np.linspace(a, b, n + 1)
    y = f(x)
    h = (b - a) / n
    return (h / 2) * (y[0] + 2 * np.sum(y[1:-1]) + y[-1])

# Método de Simpson
def simpson(f, a, b, n):
    if n % 2 == 1:  # El número de subintervalos para Simpson debe ser par
        n += 1
    x = np.linspace(a, b, n + 1)
    y = f(x)
    h = (b - a) / n
    return (h / 3) * (y[0] + 4 * np.sum(y[1:-1:2]) + 2 * np.sum(y[2:-2:2]) + y[-1])

# Método de Gauss-Legendre
def gauss_legendre(f, a, b, n):
    x, w = roots_legendre(n)
    t = 0.5 * (x + 1) * (b - a) + a  # Cambio de variable a [a, b]
    return 0.5 * (b - a) * np.sum(w * f(t))

# Función para estimar el error comparando con una solución exacta o de alta precisión
def estimate_error(approx, exact):
    return abs(approx - exact)

# Parámetros de la integración
a, b = -1, 1  # Intervalo de integración [-1, 1]
exact_value, _ = quad(f, a, b)  # Valor exacto con alta precisión usando quad

# Comparación de los métodos
n_values = [6, 15]
results = []

for n in n_values:
    # Regla del trapecio
    trap_result = trapecio(f, a, b, n)
    trap_error = estimate_error(trap_result, exact_value)

    # Regla de Simpson
    simp_result = simpson(f, a, b, n)
    simp_error = estimate_error(simp_result, exact_value)

    # Gauss-Legendre
    gauss_result = gauss_legendre(f, a, b, n)
    gauss_error = estimate_error(gauss_result, exact_value)

    # Imprimir resultados
    print(f"Resultados con {n} subintervalos/puntos:")
    print(f"Trapecio: {trap_result}, Error: {trap_error}")
    print(f"Simpson: {simp_result}, Error: {simp_error}")
    print(f"Gauss-Legendre: {gauss_result}, Error: {gauss_error}\n")

    results.append((n, trap_result, trap_error, simp_result, simp_error, gauss_result, gauss_error))
```

```python
#metodo utilizado de Gauss-legendre
import numpy as np
from scipy.special import roots_legendre

# Definir las funciones f(x) y g(x)
def f(x):
    return np.exp(x) / x

def g(x):
    return (1 - np.exp(x)) / x

# Método de Gauss-Legendre
def gauss_legendre(f, a, b, n):
    x, w = roots_legendre(n)
    t = 0.5 * (x + 1) * (b - a) + a  # Cambio de variable a [a, b]
    return 0.5 * (b - a) * np.sum(w * f(t))

# Parámetros de la integración
a, b = 0, 5  # Intervalo de integración (0, 5)

# Comparación de los métodos con 6, 15 y 20 puntos
n_values = [6, 15, 20]
functions = [f, g]
function_names = ["f(x) = e^x / x", "g(x) = (1 - e^x) / x"]

for func, name in zip(functions, function_names):
    print(f"\nIntegración de {name} en el intervalo ({a}, {b}):")

    for n in n_values:
        gauss_result = gauss_legendre(func, a, b, n)
        print(f"Gauss-Legendre con {n} puntos: {gauss_result}")

```
