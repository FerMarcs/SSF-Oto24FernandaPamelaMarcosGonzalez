```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Ejemplo de datos
datos = {
    'Genero': ['Hombre', 'Mujer', 'Hombre', 'Mujer', 'Hombre'],
    'VIQ': [112, 102, 119, 110, 123],
    'Conteo_MRI': [87845, 78212, 89876, 85432, 91000]
}
df = pd.DataFrame(datos)

# 1. Valor medio de VIQ para toda la población
media_viq = df['VIQ'].mean()
print(f"El VIQ promedio para toda la población es: {media_viq}")

# 2. Número de hombres y mujeres incluidos en el estudio
conteo_generos = df['Genero'].value_counts()
print(f"Número de hombres: {conteo_generos['Hombre']}")
print(f"Número de mujeres: {conteo_generos['Mujer']}")

# 3. Promedio de los conteos de MRI en unidades logarítmicas para hombres y mujeres
df['Log_Conteo_MRI'] = np.log(df['Conteo_MRI'])
promedio_log_mri_por_genero = df.groupby('Genero')['Log_Conteo_MRI'].mean()
print("Promedio de conteos de MRI en logaritmos por género:")
print(promedio_log_mri_por_genero)

# 4. Gráficas
plt.figure(figsize=(10, 5))

# Gráfica del promedio de VIQ
plt.subplot(1, 2, 1)
plt.bar(['Hombre', 'Mujer'], df.groupby('Genero')['VIQ'].mean(), color=['blue', 'pink'])
plt.title('VIQ promedio por género')
plt.ylabel('VIQ')

# Gráfica del promedio de los conteos logarítmicos de MRI
plt.subplot(1, 2, 2)
plt.bar(['Hombre', 'Mujer'], promedio_log_mri_por_genero, color=['blue', 'pink'])
plt.title('Promedio de log(Conteo MRI) por género')
plt.ylabel('Log(Conteo MRI)')

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)  # Para reproducibilidad
poblacion = 250
generos = np.random.choice(['Hombre', 'Mujer'], size=poblacion, p=[0.5, 0.5])

# Variables simuladas
VIQ = np.random.normal(110, 10, poblacion)  # Media 110, desviación estándar 10
Conteo_MRI = np.random.normal(85000, 5000, poblacion)  # Media 85000, desviación estándar 5000
PIQ = np.random.normal(105, 12, poblacion)  # Media 105, desviación estándar 12
FSIQ = np.random.normal(115, 8, poblacion)  # Media 115, desviación estándar 8

# Crear el DataFrame
df = pd.DataFrame({
    'Genero': generos,
    'VIQ': VIQ,
    'Conteo_MRI': Conteo_MRI,
    'PIQ': PIQ,
    'FSIQ': FSIQ
})

# Filtrar por género
hombres = df[df['Genero'] == 'Hombre']
mujeres = df[df['Genero'] == 'Mujer']

# Matriz de dispersión para hombres
sns.pairplot(hombres.drop(columns='Genero'), corner=True)
plt.suptitle("Matriz de dispersión para hombres", y=1.02)
plt.show()

# Matriz de dispersión para mujeres
sns.pairplot(mujeres.drop(columns='Genero'), corner=True, palette='husl')
plt.suptitle("Matriz de dispersión para mujeres", y=1.02)
plt.show()

# Información adicional sobre la población simulada
num_hombres = hombres.shape[0]
num_mujeres = mujeres.shape[0]
print(f"Número de hombres: {num_hombres}")
print(f"Número de mujeres: {num_mujeres}")

```
